```python
import os
import json

manually_added_pawsids = open('./manually-added.txt').read().splitlines()

for file in os.listdir('./items'):
    if not file.endswith('.json'):
        continue
    item = json.load(open(os.path.join('./items', file)))
    if 'paws_id' not in item or str(item['paws_id']) not in manually_added_pawsids:
        continue
    item['paws_id'] = None
    json.dump(item, open(os.path.join('./items', file), 'w'), indent=2)
```
