```python
import requests
import json

catalog = requests.get("http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2").json()

pcex_activities = [
    pcex_activity 
    for pcex_activity in catalog
    if pcex_activity['identity']['type'] == 'PCEX Activity'
]

api_token = open('api_token.txt').read().strip()
for pcex_activity in pcex_activities:
    print('changing to private ->', pcex_activity['identity']['title'])
    pcex_activity = requests.get(f"http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2/{pcex_activity['id']}").json()
    pcex_activity['status'] = 'private'
    response = requests.patch(
        f"http://adapt2.sis.pitt.edu/next.course-authoring/api/slc-items-api/{pcex_activity['id']}",
        json=pcex_activity,
        headers={'api-token': api_token}
    )
    print('status_code:', response.status_code)
```
