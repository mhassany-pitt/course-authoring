```python
# ------- RESTORE ITEMS -------
import os
import requests
import json

api_token = open('api_token.txt').read().strip()
items = os.listdir('items')

for index, id in enumerate(items):
    if not id.endswith('.json'):
        continue
    item = json.load(open(f'items/{id}'))
    response = requests.patch(
        f"http://adapt2.sis.pitt.edu/next.course-authoring/api/slc-items-api/{id.replace('.json', '')}",
        json=item,
        headers={'api-token': api_token}
    )
    print(f'[{index}/{len(items)}:{response.status_code}] restored item id:', id.replace('.json', ''))
```
