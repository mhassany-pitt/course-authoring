```python
# merge all items into a single file

import os
import json

items = []
for index, id in enumerate(os.listdir('items')):
    if not id.endswith('.json'):
        continue
    items.append(json.load(open(f'items/{id}')))
json.dump(items, open('items.json', 'w'), indent=2)
```
