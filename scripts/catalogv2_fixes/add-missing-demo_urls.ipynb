```python
import requests
import json

catalog = requests.get("http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2").json()
old_catalog = {item['short_name']: item['url'] for item in json.load(open('old-catalog.json'))}

api_token = open('api_token.txt').read().strip()
for index, item in enumerate(catalog):
    if "TODO:add-demo-url" in item['tags'] and not item.get('links', {}).get('demo_url'):
        print(f'{index}/{len(catalog)}:', item['id'], item['identity']['title'])
        item = requests.get(f"http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2/{item['id']}").json()
        item['links']['demo_url'] = old_catalog[item['identity']['id']]
        item['tags'].remove("TODO:add-demo-url")
        response = requests.patch(
            f"http://adapt2.sis.pitt.edu/next.course-authoring/api/slc-items-api/{item['id']}",
            json=item,
            headers={'api-token': api_token}
        )
        print('status_code:', response.status_code)
```
