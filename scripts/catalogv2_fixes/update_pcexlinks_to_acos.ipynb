```python
import os
import json

acos_pcex_links = json.load(open('./acos-pcex-links.json'))

for file in os.listdir('./items'):
    if not file.endswith('.json'):
        continue
    item = json.load(open(os.path.join('./items', file)))
    demo_url = item['links']['demo_url']
    if '/pcex-authoring/api/hub/' not in demo_url:
        continue
    wid = demo_url.split('/pcex-authoring/api/hub/')[1].split('?')[0]
    demo_url = acos_pcex_links[wid]
    item['links']['demo_url'] = demo_url
    item['delivery'] = [
        { "protocol": "SPLICE", "url": demo_url },
        { "protocol": "HTML", "url": demo_url },
        { "protocol": "PITT", "url": demo_url.replace('/html/', '/pitt/') },
    ]
    json.dump(item, open(os.path.join('./items', file), 'w'), indent=2)
```
