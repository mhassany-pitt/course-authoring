```python
import requests
import json

catalog = requests.get("http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2").json()

pcex_challenges = [
    pcex_challenge 
    for pcex_challenge in catalog
    if pcex_challenge['identity']['type'] == 'PCEX Challenge' and 'TODO:add-source-code' in pcex_challenge['tags']
]

api_token = open('api_token.txt').read().strip()
for pcex_challenge in pcex_challenges:
    print('->', pcex_challenge['identity']['title'])
    pcex_challenge = requests.get(f"http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2/{pcex_challenge['id']}").json()
    
    iframe_url = pcex_challenge['links']["demo_url"].split("index.html?")[1].split("&")
    language = iframe_url[0].split("=")[1]
    problem_set = iframe_url[1].split("=")[1]
    challenge = iframe_url[2].split("=")[1] if len(iframe_url) > 2 else None

    content = requests.get(
        f"http://pawscomp2.sis.pitt.edu/pcex/pcex_v2/data/{language}_{problem_set}.json?v=201801041411"
    ).json()
    content = (
        content["activityGoals"][0]
        if challenge is None
        else [
            g
            for g in content["activityGoals"]
            if g["fileName"].startswith(challenge)
        ][0]
    )
    
    pcex_challenge['content']['prompt'] = content["goalDescription"]
    pcex_challenge['content']['source_code'] = "\n".join(
        [l["content"] for l in sorted(content["lineList"], key=lambda l: int(l["number"]))]
    )
    pcex_challenge['tags'] = []
    
    response = requests.patch(
        f"http://adapt2.sis.pitt.edu/next.course-authoring/api/slc-items-api/{pcex_challenge['id']}",
        json=pcex_challenge,
        headers={'api-token': api_token}
    )
    print('status_code:', response.status_code)
```
