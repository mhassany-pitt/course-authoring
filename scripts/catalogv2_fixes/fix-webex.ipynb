```python
import requests
import json

catalog = requests.get("http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2").json()

webex_links = json.load(open('webex/annotated-links.json'))
webex_items = {
    key: value
    for ext in ['c', 'cpp', 'html', 'java', 'py', 'sql', 'vb', 'xml']
    for key, value in json.load(open(f'webex/annotated-{ext}.json')).items()
}

api_token = open('api_token.txt').read().strip()
for index, item in enumerate(catalog):
    if item["attribution"]['provider'] == "WebEx":
        identifier = item['identity']['id'].strip().replace('.', '_')
        if identifier == 'precedence_expression':
            identifier = 'precedence_expr'
        
        if identifier not in webex_items or identifier not in webex_links:
            continue
        
        print(f'{index}/{len(catalog)}:', f'[{identifier}]',item['identity']['title'])
        
        item = requests.get(f"http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2/{item['id']}").json()
        
        webex_desc = webex_items[identifier]['description']
        webex_code = '\n'.join([line['line'] for line in webex_items[identifier]['lines']])
        webex_link = webex_links[identifier]
        
        item['content']['prompt'] = webex_desc
        item['content']['source_code'] = webex_code
        item['links']['demo_url'] = webex_link
        item['delivery'] = [{"format": "html","url": webex_link}]
        item['tags'] = []
        
        response = requests.patch(
            f"http://adapt2.sis.pitt.edu/next.course-authoring/api/slc-items-api/{item['id']}",
            json=item,
            headers={'api-token': api_token}
        )
        print('status_code:', response.status_code)
```
