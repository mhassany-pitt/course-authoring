```python
import requests
import json

catalog = requests.get("http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2").json()

pcex_challenges = [
    pcex_challenge 
    for pcex_challenge in catalog
    if pcex_challenge['identity']['type'] == 'PCEX Challenge'
]

api_token = open('api_token.txt').read().strip()
for pcex_challenge in pcex_challenges:
    pcex_challenge = requests.get(f"http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2/{pcex_challenge['id']}").json()
    url_pattern = 'http://pawscomp2.sis.pitt.edu/pcex/index.html?'
    if pcex_challenge['links'] and pcex_challenge['links']['demo_url'] and url_pattern in pcex_challenge['links']['demo_url']:
        pcex_challenge['links']['demo_url'] = pcex_challenge['links']['demo_url']\
            .replace(url_pattern, 'http://pawscomp2.sis.pitt.edu/pcex/pcex_v2/index.html?')
        response = requests.patch(
            f"http://adapt2.sis.pitt.edu/next.course-authoring/api/slc-items-api/{pcex_challenge['id']}",
            json=pcex_challenge,
            headers={'api-token': api_token}
        )
        print('status_code:', response.status_code)
```
