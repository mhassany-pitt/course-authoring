```python
import os
import json

for filename in os.listdir('items'):
    if not filename.endswith('.json'):
        continue
    item = json.load(open(f'items/{filename}'))
    if item['attribution']['provider'] != 'PCEX':
        continue
    if item['identity']['type'] != 'CodeCompletionProblem':
        continue
    if not item['links']['demo_url'].startswith('https://acos.cs.vt.edu/html/acos-pcex/'):
        continue
    if '?index=' in item['links']['demo_url']:
        continue
    index = item['identity']['id'].split('-')[-1]
    item['links']['demo_url'] += f'?index={index}'
    for delivery in item['delivery']:
        delivery['url'] += f'?index={index}'
    json.dump(item, open(f'items/{filename}', 'w'), indent=2)
    print(f'Updated {filename}')
```
