```python
import requests

catalog = requests.get("https://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2", verify=False).json()
```

```python
import csv

pcrs_items = []
for item in catalog:
    if item['languages']['content_language'] != 'en':
        continue
    if item['attribution']['provider'] != 'PCRS':
        continue
    if "python" != item['languages']['programming_languages'][0].lower():
        continue
    pcrs_items.append({
        "id": item['identity']['id'],
        "problem_name": item['identity']['title'],
        "problem_statement": item['content']['prompt'],
    })
    
with open('pcrs_items.csv', 'w', newline='') as csvfile:
    fieldnames = ['id', 'problem_name', 'problem_statement']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for item in pcrs_items:
        writer.writerow(item)
```

```python
import csv

parsons_items = []
for item in catalog:
    if item['languages']['content_language'] != 'en':
        continue
    if item['attribution']['provider'] != 'jsParsons':
        continue
    if "python" != item['languages']['programming_languages'][0].lower():
        continue
    item = requests.get(f"https://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2/{item['id']}", verify=False).json()
    parsons_items.append({
        "id": item['identity']['id'],
        "problem_name": item['identity']['title'],
        "problem_statement": item['content']['prompt'],
        "solution_code": item['content']['source_code'],
    })
    
with open('parsons_items.csv', 'w', newline='') as csvfile:
    fieldnames = ['id', 'problem_name', 'problem_statement', 'solution_code']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for item in parsons_items:
        writer.writerow(item)
```
