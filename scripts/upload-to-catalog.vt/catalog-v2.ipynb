```python
import requests

catalog = requests.get("http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2").json()
```

```python
import os
import json

os.makedirs("paws-catalog", exist_ok=True)

for index, item in enumerate(catalog):
    if os.path.exists(f"paws-catalog/{item['id']}.json"):
        print(f"{index}/{len(catalog)} {item['id']} already exists, skipping.")
        continue
    
    item = requests.get(f"http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2/{item['id']}").json()
    with open(f"paws-catalog/{item['id']}.json", "w") as f:
        json.dump(item, f, indent=2)
    print(f"{index}/{len(catalog)} {item['id']} downloaded.")
```
