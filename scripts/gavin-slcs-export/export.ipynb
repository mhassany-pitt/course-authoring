```python
muntasir = pl.read_csv('Muntasir_Gavin -- 2025-12-17 __ PCEX Dataset - pcex_final+kcs.csv')
muntasir = muntasir.filter(pl.col('item_language') == 'PYTHON')
muntasir
```

```python
# replace
# set_name: py_nestedloops_print_a_square_star_pattern --> py_nestedloops_print_a_right_square_star_pattern
# item_name: print_a_square_star_pattern_ch --> print_a_right_square_star_pattern_ch
muntasir = muntasir.with_columns([
    pl.when(pl.col('set_name') == 'py_nestedloops_print_a_square_star_pattern').then(pl.lit('py_nestedloops_print_a_right_square_star_pattern')).otherwise(pl.col('set_name')).alias('set_name'),
    pl.when(pl.col('item_name') == 'print_a_square_star_pattern_ch').then(pl.lit('print_a_right_square_star_pattern_ch')).otherwise(pl.col('item_name')).alias('item_name')
])
```

```python
import requests

catalog = {
    item['identity']['id']: item
    for item in requests.get('http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2').json()
}

kc_sets = {}
for row in muntasir.iter_rows(named=True):
    item = None
    if row['item_type'] == 'PCEX Set' and row['set_name'] in catalog:
        item = catalog[row['set_name']]
    elif row['item_type'] == 'PCEX Challenge' and row['item_name'] in catalog:
        item = catalog[row['item_name']]
    print(row['item_type'], row['set_name'], row['item_name'], '->', item['identity']['id'] if item else 'NOT FOUND')
    item = requests.get(f"http://adapt2.sis.pitt.edu/next.course-authoring/api/catalog-v2/{item['id']}").json()
    kc_sets[f"{row['item_type']}::{row['set_name']}::{row['item_name']}"] = item['classification']['knowledge_components'] \
        if 'classification' in item and 'knowledge_components' in item['classification'] \
        else {}
```

```python
classifications = set()
for key, kc_set in kc_sets.items():
    classifications.update(kc_set.keys())
classifications
```

```python
def get_kcs(row, classification):
    key = f"{row['item_type']}::{row['set_name']}::{row['item_name']}"
    return kc_sets[key][classification]['concepts'] if classification in kc_sets[key] else []

pl.Config.set_tbl_rows(10)
muntasir = muntasir.with_columns([
    pl.struct(pl.all()).map_elements(lambda row: get_kcs(row, 'Arun added 08.12.2025 Arun Parser v0.1'), return_dtype=pl.List(pl.Utf8)).alias('Arun added 08.12.2025 Arun Parser v0.1'),
    pl.struct(pl.all()).map_elements(lambda row: get_kcs(row, 'Arun added 08.15.2025 Arun Parser v0.1'), return_dtype=pl.List(pl.Utf8)).alias('Arun added 08.15.2025 Arun Parser v0.1'),
    pl.struct(pl.all()).map_elements(lambda row: get_kcs(row, 'Arun parser v0.1'), return_dtype=pl.List(pl.Utf8)).alias('Arun parser v0.1'),
    pl.struct(pl.all()).map_elements(lambda row: get_kcs(row, 'Rafaella added 09.03.2025 alice_edu_py_ontology'), return_dtype=pl.List(pl.Utf8)).alias('Rafaella added 09.03.2025 alice_edu_py_ontology'),
    pl.struct(pl.all()).map_elements(lambda row: get_kcs(row, 'alice_edu_py_ontology expert'), return_dtype=pl.List(pl.Utf8)).alias('alice_edu_py_ontology expert'),
    pl.struct(pl.all()).map_elements(lambda row: get_kcs(row, 'parser'), return_dtype=pl.List(pl.Utf8)).alias('parser'),
    pl.struct(pl.all()).map_elements(lambda row: get_kcs(row, 'unspecified'), return_dtype=pl.List(pl.Utf8)).alias('unspecified'),
])
muntasir
```

```python
muntasir.with_columns([
    pl.struct(['Arun added 08.12.2025 Arun Parser v0.1','Arun added 08.15.2025 Arun Parser v0.1','Arun parser v0.1','parser']).map_elements(
        lambda row: list(set([
            *row['Arun added 08.12.2025 Arun Parser v0.1'],
            *row['Arun added 08.15.2025 Arun Parser v0.1'],
            *row['Arun parser v0.1'],
            *row['parser'],
        ])), return_dtype=pl.List(pl.Utf8)
    ).alias('Arun added parser')
]).drop([
    'Arun added 08.12.2025 Arun Parser v0.1',
    'Arun added 08.15.2025 Arun Parser v0.1',
    'Arun parser v0.1',
    'parser',
]).rename({
    'Rafaella added 09.03.2025 alice_edu_py_ontology': 'alice_edu_py_ontology',
    'alice_edu_py_ontology expert': 'alice_edu_py_ontology expert',
    'unspecified': 'unspecified',
    'Arun added parser': 'parser',
}).with_columns(
    pl.col('item_kcs').map_elements(lambda v: ','.join(sorted(list(set(v.split(','))))), pl.Utf8).alias('item_kcs'),
    pl.col('parser').map_elements(lambda v: ','.join(sorted(list(set(v)))), pl.Utf8).alias('parser_kcs'),
    pl.col('alice_edu_py_ontology').map_elements(lambda v: ','.join(sorted(list(set(v)))), pl.Utf8).alias('alice_edu_py_ontology_kcs'),
    pl.col('alice_edu_py_ontology expert').map_elements(lambda v: ','.join(sorted(list(set(v)))), pl.Utf8).alias('alice_edu_py_ontology_expert_kcs'),
    pl.col('unspecified').map_elements(lambda v: ','.join(sorted(list(set(v)))), pl.Utf8).alias('unspecified_kcs'),
)[[
    'item_language',
    'set_name',
    'item_name',
    'item_type',
    'item_statement',
    'item_code',
    'item_url',
    'item_kcs',
    'python_parser_kcs',
    'parser_kcs',
    'alice_edu_py_ontology_kcs',
    'alice_edu_py_ontology_expert_kcs',
    'unspecified_kcs',
]] #.to_pandas().to_clipboard(index=False)
```
