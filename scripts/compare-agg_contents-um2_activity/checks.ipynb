```python
import polars as pl

ent_activity = pl.read_csv('ent_activity_202601121152.csv')\
    .filter(pl.col('AppID').is_in([45, 46, 47]))\
    .filter(~pl.col('Description').str.contains('PCEX Line - '))\
    .filter(~pl.col('URI').str.contains('http://pawscomp2.sis.pitt.edu/pcex-authoring/preview/index.html'))
print(ent_activity.shape)
ent_activity.head(5)
```

```python
ent_content = pl.read_csv('ent_content_202601121152.csv')\
    .filter(pl.col('provider_id').is_in(['pcex', 'pcex_ch']))\
    .filter(~pl.col('url').str.contains('http://pawscomp2.sis.pitt.edu/pcex-authoring/preview/index.html'))
print(ent_content.shape)
ent_content.head(5)
```
