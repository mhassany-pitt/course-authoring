```python
for item in webex:
    if item["domain"] is None:
        item["domain"] = "c"

for lang in ["c", "cpp", "html", "java", "py", "sql", "vb", "xml"]:
    annotated = dict()
    for item in webex:
        if item["rdfID"] in annotated:  # or item["rdfID"] not in filter:
            continue
        if item["domain"] == lang:
            annotated[item["rdfID"].strip()] = {
                "name": item["Name"].strip(),
                "description": item["Description"] or "",
                "language": item["domain"],
                "author": item["Author"],
                "lines": [
                    {"line": line["Code"], "comment": line["Comment"]}
                    for line in webex
                    if line["DissectionID"] == item["DissectionID"]
                ],
            }

    json.dump(annotated, open(f"./annotated-{lang}.json", "w"), indent=2)
    print(len(annotated))
```
